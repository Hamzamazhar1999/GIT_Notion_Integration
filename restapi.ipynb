### Import libraries and token for REST GitHub API

```python
import requests
from github_pat import pat
import base64

TOKEN = pat
HEADERS = {"Authorization": f"Bearer {TOKEN}"}
#base url
BASE_URL = "https://api.github.com/repos/geekcomputers/Python"
```

### Initial Fetch

```python
response = requests.get(BASE_URL, headers=HEADERS)
print(response.json())
```

### List out the contents

```python
url = f"{BASE_URL}/contents/"
response = requests.get(url, headers=HEADERS)
print(response.json())
for item in response.json():
    print(item["name"], "-", item["size"])
```

### Get content as base64 encoded

```python
file_path = "JARVIS/JARVIS_2.0.py"
url = f"{BASE_URL}/contents/{file_path}"
response = requests.get(url, headers=HEADERS)
content = response.json()["content"]
decoded_content = base64.b64decode(content).decode("utf-8")
print(decoded_content)
```

```python
from openai import OpenAI
from apikey import gpt_api


# setup api client
client = OpenAI(api_key=gpt_api)

# make api call
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Expert software engineer summarization for code, with explanation"},
        {"role": "user", "content": decoded_content}
    ], 
    temperature = 0.7
)

# extract response
response_string = response.choices[0].message.content
# print(response_string)
```
