### Import libraries and token for REST GitHub API

```python
import requests
from github_pat import pat
import base64

TOKEN = pat
HEADERS = {"Authorization": f"Bearer {TOKEN}"}
#base url
BASE_URL = "https://api.github.com/repos/deepseek-ai/DeepSeek-V3"

```

### Initial Fetch

```python
response = requests.get(BASE_URL, headers=HEADERS)
print(response.json())
```

### List out the contents

```python
url = f"{BASE_URL}/contents/"
response = requests.get(url, headers=HEADERS)
for item in response.json():
    print(item["name"], "-", item["type"])
```

### Get content as base64 encoded

```python
file_path = "README.md"
url = f"{BASE_URL}/contents/{file_path}"
response = requests.get(url, headers=HEADERS)
content = response.json()["content"]
decoded_content = base64.b64decode(content).decode("utf-8")
print(decoded_content)
```
